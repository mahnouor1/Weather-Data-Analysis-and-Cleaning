```python
import pandas as pd  #pandas library is used to display data in tabular form
df = pd.read_csv("Weather Test Data.csv")   #this function is used to read data from the file
df
```

```python
df.describe()  #this func is used to define statistical formula of numerical columns
```

# Data Cleaning

```python
null_values = df.isnull().sum() #isnull func used to show null values in columns, .sum() displays the sum of all null values from columns
null_values
```

```python
#fill NaN with mean of that column
import numpy as np  #numpy used to handle numeric data

# Calculate the mean of each column
column_means = df.select_dtypes(include=np.number).mean() # Only calculate the mean of numeric columns
# Fill NaN values with the mean of the respective column
df.fillna(column_means, inplace=True)  #inplace=true modifies the original object (like a DataFrame or Series) directly without returning a new copy

column_means
```

```python
# drop null values NaN values

df.dropna(inplace=True)
df
```

```python
df.shape #remaining rows are left now
```

```python
#drop unimportant columns

#WindGustDir, WindDir9am, WindDir3pm are unimportant columns because it helps focus on the more relevant data

df = df.drop(['WindGustDir', 'WindDir9am', 'WindDir3pm'], axis=1, errors='ignore') #errors='ignore' will skip any columns that are not found
df
```

```python
#get duplicate samples from df

duplicate_samples = df[df.duplicated(keep=False)]  #Marks all occurrences of duplicate rows as duplicates.
print(duplicate_samples)
```

```python
df.columns #display all columns
```

```python
#renaming columns for more clarification
df.rename(columns={
    'row ID': 'Row ID',
    'Location': 'Locations',
    'MinTemp': 'Minimum Temperature',
    'MaxTemp': 'Maximum Temperature',
    'Evaporation': 'Evaporations',
    'WindGustSpeed': 'Wind Speed',
    'WindSpeed9am': 'Wind Speed at 9am',
    'RainTomorrow': 'Rain Tomorrow',
    'RainToday': 'Rain Today'
}, inplace=True)
df.head()
```

```python
#The code counts how many times each unique value appears in the 'Wind Speed' column of the DataFrame df_new,

type_counts = df['Wind Speed'].value_counts()
type_counts
```

# Pie Chart Of RainFall vs Sunshine

```python
import matplotlib.pyplot as plt

# Step 1: Select the 'Rainfall' and 'Sunshine' columns and sum their values
rainfall_sum = df['Rainfall'].sum() #.sum() giving you the total percentage of rainfall across the dataset.
sunshine_sum = df['Sunshine'].sum()

# Step 2: Create a list with the sums
values = [rainfall_sum, sunshine_sum]

# Step 3: Create labels for the pie chart
labels = ['Rainfall', 'Sunshine']

# Step 4: Define an explode to slightly separate one slice for emphasis
explode = [0, 0.1]  # Only explode the second slice (Sunshine)

# Step 5: Plot the pie chart
plt.figure(figsize=(5, 5))
plt.pie(values, labels=labels, explode=explode, autopct='%1.2f%%', shadow=True)
plt.title("Comparison of Rainfall vs Sunshine")
plt.show()
```

# Bar Graph of Highest Wind Speed at 9am by Wind Speed

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Calculate the maximum Wind Speed at 9am for each Wind speed
wind_speed = df.groupby('Wind Speed at 9am')['Wind Speed'].max() # groups the DataFrame by the Wind Speed at 9am column and
#calculates the maximum value of the Wind Speed column for each group.

# Step 2: Sort the results by Wind Speed in descending order
wind_speed = wind_speed.sort_values(ascending=False) #sorts the maximum wind speeds in descending order, so the highest values appear first

# Step 3: Create the bar plot
plt.figure(figsize=(15, 7))

sns.barplot(x=wind_speed.index, y=wind_speed.values, palette='coolwarm')
plt.xticks(rotation=30, ha="right")
plt.title("Highest Wind Speed at 9am by Wind Speed")
plt.xlabel("Wind Speed at 9am")
plt.ylabel("Highest Wind Speed")
plt.tight_layout()
plt.show()
```

# Displaying Available Outliers in the Data




```python
# create box plots and distribution

import matplotlib.pyplot as plt
import seaborn as sns

# Loop through numerical columns and create box plots and histograms
for column in df.select_dtypes(include=['number']).columns:
  plt.figure(figsize=(12, 6))

  # Box plot
  plt.subplot(1, 2, 1)
  sns.boxplot(x=df[column])
  plt.title(f'Box Plot of {column}')

  # Histogram
  plt.subplot(1, 2, 2)
  plt.hist(df[column], bins=20)
  plt.title(f'Distribution of {column}')

  plt.show()
```

# Removing Outliers from Wind Speed

```python
#remove outlier from Wind Speed

# Calculate the IQR for 'Wind Speed'
Q1 = df['Wind Speed'].quantile(0.25)
Q3 = df['Wind Speed'].quantile(0.75)
IQR = Q3 - Q1

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")

# Remove outliers based on the bounds
df_no_outliers = df[(df['Wind Speed'] >= lower_bound) & (df['Wind Speed'] <= upper_bound)]

# Print the number of rows before and after outlier removal
print(f"Number of rows before outlier removal: {df.shape[0]}")
print(f"Number of rows after outlier removal: {df_no_outliers.shape[0]}")
```

```python
# prompt: plot box plot and distribution for df_no_outliers Wind Speed

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

# Box plot
plt.subplot(1, 2, 1)
sns.boxplot(x=df_no_outliers['Wind Speed'])
plt.title(f'Box Plot of Wind Speed (No Outliers)')

# Histogram
plt.subplot(1, 2, 2)
plt.hist(df_no_outliers['Wind Speed'], bins=20)
plt.title(f'Distribution of Wind Speed (No Outliers)')

plt.show()
```
